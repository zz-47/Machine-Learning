```python
# This is a python NB to Visualize the cost function...

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

```

```python
# providing data set...

X = np.linspace(0 , 10 , 50 )

true_w = 2 # variable
true_b = 1 # constant

# nosie in data set to make it less perfect yet predictable

noise = np.random.randn(50) * 2 

y = true_w * X + true_b + noise # i.e  2X + 1 + noise

# visualizing the data set...

plt.scatter( X, y , label = "Cars")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Generated Data-Set")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Creating Cost function / J(x) / sole purpose of this project...

# using mean squared error .i,e (y - yhat)^2 ... m Number of terms...

def cost_func(w ,b ,X ,y ): 
    m = len(X) # num of terms

    pred = w * X + b  # i.e 2X + 1

    # now using formula : J(w,b) = 1/2m = 1∑m​ (y^​(i)−y(i))2

    cost = (1/(2*m)) * np.sum( (pred - y) ** 2 )
    
    return cost
```

```python
w_val  = np.linspace(0, 4, 100)

# calculating cost function for each value of w

cost_val = [cost_func(w ,0 ,X ,y ) for w in w_val]

print("Cost Values Formatted : ")

for w, cost in zip(w_val , cost_val):
    print(f'w = {w:.2f} --> Cost = {cost :.2f}')

plt.plot(w_val, cost_val)
plt.xlabel("Engine Liters (cc)")
plt.ylabel("Cost (1000 $)")
plt.title("Cost function b/w [ 'w' and 'b' ]")
plt.grid(True)
plt.show()
```

```python
# creating a 3-D visualization of the cost function...

w_range = np.linspace(0, 4, 50)
b_range = np.linspace(-5 ,5 ,50)

W, B = np.meshgrid(w_range, b_range)

# creating a 3rd dimensin Z (.i.e x,y,z place / space)

Z = np.zeros_like(W)

for i in range(W.shape[0]): 
    for j in range(W.shape[1]):
        Z[i,j] = cost_func(W[i,j], B[i,j], X, y)


# plotting 3D surface

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(W, B, Z, cmap = 'viridis', edgecolor = 'None')
ax.set_xlabel('Engine Capacity (cc)') 
ax.set_ylabel('bias (b)')
ax.set_zlabel('cost ( 1000 $ )')
ax.set_title('Cost Function Surface')
plt.show()
```
