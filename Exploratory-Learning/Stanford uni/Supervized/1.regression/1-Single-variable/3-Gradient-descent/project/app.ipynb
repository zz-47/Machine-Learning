```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating data set

X = np.array([1,2,3,4,5],dtype = float)
y = np.array([500,600,700,800,900],dtype = float)
```

```python
# innit paramters.
w, b = 0, 0

# adjust learning rate
lr = 0.0001

# adjust iterations
iterate = 1000
```

```python
# starting Batch - Gradient - Descent

for i in range(iterate):
    
    pred_y = w * X + b 

    diff = pred_y - y

#    w = w - lr * ( 1 / n ​∑( y_pred​ − y ) x ) 
#    b = b - lr * ( 1 / n ​∑( y_pred​ − y ) ) 

    w -= lr * (diff * X).mean()
    b -= lr * diff.mean()
```

```python
# final result 

print(f' Final model : y = {w:.2f} X + {b:.2f}')
```

```python
# plotting the graph

plt.scatter(X,y, color = 'blue', label = 'predected prices ' )
plt.plot(X, w*X + b, color = 'red', label = ' Range Prices ' )
plt.xlabel('House Size ( 1000 sqft)')
plt.ylabel("Price ( 1000 $)") 
plt.legend()
plt.title("Linear Regression : House Price Prediction")
plt.grid()
plt.show()
```
